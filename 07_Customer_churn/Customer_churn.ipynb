## Исследование задачи

Создадим функцию построения графика ROC-кривой и других метрик:

```python
def roc_plot(model, X_train=X_train, y_train=y_train):

    model.fit(X_train, y_train)
    probabilities_valid = model.predict_proba(X_valid)
    probabilities_one_valid = probabilities_valid[:, 1]

    fpr, tpr, thresholds = roc_curve(y_valid, probabilities_one_valid)

    plt.figure(figsize=(4.0, 4.0))

    plt.plot(fpr, tpr, label='data')
    # ROC-кривая случайной модели (выглядит как прямая)
    plt.plot([0, 1], [0, 1], linestyle='--', label='random classifier')


    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')
    plt.legend()
    plt.grid()
    plt.show()

    print('model:', model)
    print(f'AUC-ROC: {roc_auc_score(y_valid, probabilities_one_valid):0.4f} \
    ,   F1_score: {f1_score(y_valid, model.predict(X_valid)):0.4f}')
    print('confusion_matrix:')
    print(confusion_matrix(y_valid, model.predict(X_valid)))
```

### Алгоритм классификации — _DecisionTreeClassifier_

```python
#Модель с базовыми гиперпараметрами:
model = DecisionTreeClassifier(random_state=12345)
roc_plot(model)
```

Попробуем улучшить модель, создадим функцию:

```python
def search_best_dtc(X_train=X_train, y_train=y_train, class_weight=None):

    best_model = None
    best_result = 0

    for depth in range(1, 21):
        for mss in range(2, 120, 5):
                for msl in range(1, 12):
                    model = DecisionTreeClassifier(random_state=12345, max_depth=depth,  \
                                 min_samples_split=mss, min_samples_leaf=msl, \
                                                   class_weight=class_weight)
                    model.fit(X_train, y_train)

                    predictions = model.predict(X_valid)

                    result = f1_score(y_valid, predictions)
                    if result > best_result:
                        best_model = model
                        best_result = result

    print("Лучшая модель:", best_model)
    print(f"F1 лучшей модели: {best_result:0.4f}")
    return best_model
```

```python
def search_best_rfc(X_train=X_train, y_train=y_train, class_weight=None):

    best_model = None
    best_result = 0

    for depth in range(3, 25, 3):
        for estimator in range(8, 100, 4):
            for msl in range(2, 15, 2):
                model = RandomForestClassifier(random_state=12345, max_depth=depth,
                                               n_estimators=estimator, min_samples_leaf=msl,
                                              class_weight=class_weight)
                model.fit(X_train, y_train)
                result = f1_score(y_valid, model.predict(X_valid))
                if result > best_result:
                    best_model = model
                    best_result = result

    print("Наилучшая модель:", best_model)
    print(f"F1 наилучшей модели: {best_result:0.4f}")
    return best_model
```

### Алгоритм классификации — _LogisticRegression_

```python
def search_best_lr(X_train=X_train, y_train=y_train, class_weight=None):
    best_model = None
    best_result = 0

    classifier__penalty = ['l1', 'l2']
    classifier__C = np.logspace(-4, 4, 20)
    classifier__solver = ['liblinear', 'saga']

    for penalty in classifier__penalty:
        for c in classifier__C:
            for solver in classifier__solver:
                model = LogisticRegression(random_state=12345, penalty=penalty, \
                                              C=c, solver=solver, class_weight=class_weight)
                model.fit(X_train, y_train)
                result = f1_score(y_valid, model.predict(X_valid))
                if result > best_result:
                    best_model = model
                    best_result = result

    print("Наилучшая модель:", best_model)
    print(f"F1 наилучшей модели: {best_result:0.4f}")
    return best_model
```

### Техника **upsampling**

Подготовим данные для обучения:

```python
def upsample(X_train, y_train, repeat):
    features_zeros = X_train[y_train == 0]
    features_ones = X_train[y_train == 1]
    target_zeros = y_train[y_train == 0]
    target_ones = y_train[y_train == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled, random_state=12345)
    return features_upsampled, target_upsampled
```

```python
X_upsampled, y_upsampled = upsample(X_train, y_train, 4)

print(X_upsampled.shape)
print(y_upsampled.shape)
```
